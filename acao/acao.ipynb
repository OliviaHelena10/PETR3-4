```python
# BIBLIOTECAS
import yfinance as yf
import investpy as inv
```

```python
# Importar ações do Brasil
br = inv.stocks.get_stocks(country = 'brazil')
display(br)
```

```python
# Capturar a nomenclatura das açôes
br['symbol']
```

```python
# retirar oq não é ação
# parte opicional para este caso - exceto a adição do termo ".SA"

carteira = []
for a in br['symbol'] :
    if len(a) <= 5:
        carteira.append(a + '.SA') # Adicionar .SA para que o yfinance consiga pescar as ações

display(carteira)
```

```python
# selecionar apenas as ações que serão utilizadas
acoes = ['PETR3.SA', 'PETR4.SA']

tabela = yf.download(acoes, start = "2019-01-01", end = "2024-12-03")['Adj Close']
display(tabela)
```

```python
# Construção do Gráfico
import plotly.express as px

graf = px.line(tabela, y = ['PETR3.SA', 'PETR4.SA'])
display(graf)
```
